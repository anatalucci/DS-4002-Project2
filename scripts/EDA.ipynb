```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/cleaned_data.csv") # upload dataset 
```

```python
# plot of top ten states by total customers affected 
state_impact = df.groupby('state')['mean_customers'].sum().sort_values(ascending=False).head(10) 
# group all events by state, sum mean_customers per state to find total customers affected per state

plt.figure(figsize=(6, 4))
state_impact.plot(kind='barh', color='skyblue')
plt.title('Top 10 States by Total Customers Affected')
plt.xlabel('Total Mean Customers')
plt.ylabel('State')
plt.gca().invert_yaxis()
plt.savefig("../outputs/EDA/total_customers_affected_by_state.jpg", format='jpg', dpi=100) 
plt.close()
```

```python
# convert time columns to datetime objects
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

# extract useful time features
df['month'] = df['start_time'].dt.month_name()
df['hour'] = df['start_time'].dt.hour
df['day_of_week'] = df['start_time'].dt.day_name()

# set month order 
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(4, 3, figsize=(15, 18))
axes = axes.flatten()

for i, month in enumerate(month_order):
    # filter data for the specific month
    month_data = df[df['month'] == month]
    
    # get the top 10 states for that month
    top_states = month_data['state'].value_counts().head(10)
    
    # plot on the specific subplot axis
    sns.barplot(x=top_states.values, y=top_states.index, ax=axes[i])
    
    axes[i].set_title(f'Top 10 States: {month}', fontsize=16, fontweight='bold')
    axes[i].set_xlabel('Number of Outages')
    axes[i].set_ylabel('')

# adjust layout to prevent overlapping
plt.tight_layout()
plt.savefig("../outputs/EDA/top_ten_states_by_month.jpg", format='jpg', dpi=100) 
plt.close()
```

```python
# plot of frequency of outages by month 
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='month', order=month_order)
plt.title('Frequency of Outages by Month')
plt.xticks(rotation=45)
plt.savefig("../outputs/EDA/frequency_of_outages_by_month.jpg", format='jpg', dpi=100) 
plt.close() 
```

```python
plt.figure(figsize=(10, 8))

# count the frequency of each Event Type and sort them
event_counts = df['Event Type'].value_counts()

sns.barplot(x=event_counts.values, y=event_counts.index)
plt.title('Distribution of Outages by Event Type', fontsize=18)
plt.xlabel('Total Number of Records', fontsize=14)
plt.ylabel('Event Type', fontsize=14)

for i, v in enumerate(event_counts.values):
    plt.text(v + 50, i, str(v), color='black', va='center', fontweight='bold')
    
plt.savefig("../outputs/EDA/distribution_of_outages_by_type.jpg", format='jpg', dpi=100) 
plt.close()
```
